# Algoritmo para transformar las aceleraciones del archivo excel a XML

```python
d=pd.read_excel("https://raw.githubusercontent.com/NoeJA/DatosP/main/xmltest_v1.xlsx",sheet_name="ACEL TRANS")
d.columns = d.columns.str.replace(' ', '') 
```

```python
d.head()
```

```python
#Tomamos los títulos
l=[]
names=[]
for i in range(len(d.columns)):
  if "Unnamed"  not in d.columns[i]:
    l.append(i)
    names.append(d.columns[i])
```

```python
new_header = d.iloc[0]
```

```python
#Eliminamos englones con NAs
d = d.rename(columns = new_header)
d=d.drop([0,1])
```

```python
l=[]
import numpy as np
for i in range(len(d.iloc[0])):
  if pd.isna(d.iloc[0][i]):
    l.append(i)
```

Se eliminan caracteres que el convertidor a xml no puede utilizar como encabezados

```python
d.columns = d.columns.str.replace('[#,@,(,[,&,%,\],),:,.,=,/]', '')
d.columns = d.columns.str.replace(' ', '')
```

```python
d
```

```python
import pandas as pd
from lxml import etree as et
data = d
col=list(data.columns)
```

```python
names[2]=names[2].replace('(', '')
names[2]=names[2].replace(')', '')#Eliminamos caracteres que no pueden ser usados como encabezados
```

```python
names
```

```python
auxl=[0]+list(map(lambda x: x+1,l))+[len(col)]#Creamos una lista para saber donde iniciar y donde finalizar
```

```python
root = et.Element('Aceleraciones')
for k in range(len(names)):
  for row in data.iterrows():#Esto recorre los renglones del Excel
      root_tags = et.SubElement(root, names[k]) 
      #etiquetas por cada renglón
      for i in range(auxl[k],auxl[k+1]):
        if i not in l and type(row[1][col[i]])!=pd.core.series.Series:#Revisamos que los datos sean datos duros
          Column_heading_ = et.SubElement(root_tags, col[i].lower())#[:(len(col[i]))-1])
          ###Valores de cada columna
          Column_heading_.text = str(row[1][col[i]])
        elif i not in l and type(row[1][col[i]])==pd.core.series.Series:#si no tenemos dastos duros(serie) tomamos el valor como tal de la serie
          Column_heading_ = et.SubElement(root_tags, col[i].lower())#[:(len(col[i]))-1])
          ###Valores de cada columna
          Column_heading_.text = str(row[1][col[i]][0])
# Se escribe el XML
tree = et.ElementTree(root) #==> The variable tree is to hold all the values of "root"
tree.write('output.xml', encoding="utf-8")
```

```python
x = et.parse("output.xml")
pretty_xml1 = et.tostring(x, pretty_print=True, encoding=str)
```

```python
pretty_xml1
```
